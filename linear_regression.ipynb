 1. Import required libraries





```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
```

   2. Load titanic data set

```python
df = pd.read_csv("titanic.csv")
df.head(5)
```

  3.Data preprocessing

```python
df = df[['Age', 'Fare']].dropna()
```

4. split data into X and Y

```python
X = df[['Age']]     # Independent variable
y = df['Fare']      # Dependent variable
```

5. Train test split

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=120)
```

6. Apply Linear Regression

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

7. Prediction And Matric

```python
y_pred = model.predict(X_test)

# Mean Squared Error
mse = mean_squared_error(y_test, y_pred)
print("Mean Squared Error:", mse)

# R² Score
r2 = r2_score(y_test, y_pred)
print("R2 Score (Mean Score):", r2)

# Intercept and Coefficient
print("Intercept:", model.intercept_)
print("Coefficient (Slope):", model.coef_[0])
```

8. Plot the regression line

```python
plt.scatter(X_test, y_test, color='blue', label='Actual Fare')
plt.plot(X_test, y_pred, color='red', linewidth=2, label='Predicted Fare')
plt.xlabel("Age")
plt.ylabel("Fare")
plt.title("Linear Regression on Titanic Dataset")
plt.legend()
plt.grid(True)
plt.show()
```
